```python
# Salary Data Insights: A Detailed Exploratory Analysis
# A part of Big Data Analysis

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading Dataset

salary = pd.read_csv(r'C:\\Users\\user\\Desktop\\DATA ANALYST\\Python\\Salary Dataset.csv')
```

```python
salary.head()
```

```python
# Adding new column of 'Age_Group'

salary['age_group'] = pd.cut(salary['Age'], [10, 18, 30, 50, 80],
                             labels=['child', 'youth', 'middle aged', 'old'])
```

```python
# Visualize the null values in Heatmap

sns.heatmap(salary.isnull(),yticklabels=False, cbar=False, cmap='viridis')
```

```python
salary.columns
```

```python
# Filling null values of all columns
# Filling nan values 'age' column

salary.Age.fillna(salary['Age'].mean(), inplace=True)
```

```python
# Filling nan values 'Year of Experience' column

salary['Years of Experience'].fillna(salary['Years of Experience'].mode()[0], inplace=True)
```

```python
# Filling nan values 'Salary' column

salary['Salary'].fillna(salary['Salary'].mean(), inplace=True)
```

```python
# Filling nan values 'Gender' column

salary['Gender'].fillna(salary['Gender'].mode()[0], inplace=True)
```

```python
# Filling nan values 'Education level' column

salary['Education Level'].fillna(salary['Education Level'].mode()[0], inplace=True)
```

```python
# Filling nan values 'Job Title' column

salary['Job Title'].fillna(salary['Job Title'].mode()[0], inplace=True)
```

```python
# Boxplot for checking outliers

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have imported the necessary libraries and loaded your data

# Create a boxplot of Salary distribution
sns.boxplot(data=salary, y='Salary')

# Display the plot
plt.show()
```

```python
# Boxplot b/w Salary & Education Level

sns.barplot(data=salary, y='Salary', x='Education Level', hue='Gender')
```

```python
# Data Distribution

sns.histplot(data=salary, x='Salary', bins=20)
```

```python
# Scatterplot 
# Create a scatter plot of Age vs Salary

sns.scatterplot(data=salary, x='Age', y='Salary')
```

```python
sns.scatterplot(data=salary, x='Years of Experience', y='Salary')
plt.xticks(rotation=0)
plt.show()
```

```python
salary.groupby('age_group')['Salary'].mean().plot(kind='bar')
```

```python
sns.pairplot(salary)
```

```python
# pairplot for complete overview

sns.pairplot(salary, vars=['Age', 'Gender', 'Education Level', 'Years of Experience', 'Salary'])
```

```python
# Stacked barchart/plot

salary.groupby(['Years of Experience', 'Gender'])['Salary'].mean().unstack().plot(kind='bar')
```

```python
# Swarm plot

sns.swarmplot(data=salary, x='Education Level', y='Age', hue='Gender') # swarm plot gives bar
```

```python
# Pivot table (city, cases)

table=pd.pivot_table(salary, values='Salary', index='Education Level', columns='Gender', aggfunc='mean')
table
```

```python
# Select numerical columns for correlation
numerical_columns = ['Age', 'Years of Experience', 'Salary']

# Create a correlation matrix
correlation_matrix = salary[numerical_columns].corr()

# set the figure size (optional)
plt.figure(figsize=(10, 8))

# create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)  # Fixed typo here

# show the plot
plt.show()
```
